```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.misc import imread, imshow
from pylab import *

import copy
```

```python
file_path_1 =  "hw4_1_1_isocontour.png"
file_path_2 =  "hw4_1_2_isocontour.jpg"
```

```python
def isocontour(file_path, number):
    # 读入图片,转换成灰度图格式,再转化为数组格式
    image = np.array(Image.open(file_path).convert("L")).astype(np.uint8)
    x, y = image.shape

    # !!!Notice:"=" 不起作用，因为只会赋值地址，所以用copy
    image_new_array = copy.deepcopy(image)

    for i in range(x):
        for j in range(y):
            # 将等高线用黑色线画出
            if image[i, j] == number:
                image_new_array[i, j] = 0

    #作图
    plt.figure(figsize=(10, 10))
    
    plt.subplot(2, 1, 1)
    plt.title('origin image')
    plt.imshow(image, cmap="gray")
    
    plt.subplot(2, 1, 2)
    plt.title('isocontour ' + str(number))
    plt.imshow(image_new_array, cmap="gray")
    #存储
    plt.savefig('1_isocontour.png')
```

```python
#给定灰度值，画出一条等值线
isocontour(file_path_1, 200)
```

```python
#给定灰度值，画出一条等值线
isocontour(file_path_2, 150)
```
