```python
import pandas as pd
import numpy as np
from pyecharts import Timeline
from pyecharts import Map
```

```python
gdp = pd.read_csv('./data/GDP20.csv',encoding='utf-8')
gdp.head()
```

```python
#实例化Timeline
timeline = Timeline(is_auto_play=1,timeline_bottom=0)
```

```python
col_names = gdp.columns.values.tolist()
max_gdp = 0
min_gdp = np.inf

#GDP取值范围
for col_name in col_names[1:]:
    colmax = max(gdp[col_name])
    colmin = min(gdp[col_name])
    if colmax > max_gdp:
        max_gdp = colmax
    if colmin < min_gdp:
        min_gdp = colmin
```

```python
for col_name in col_names[1:]:
    value = list(gdp[col_name])
    attr = list(gdp.country)
    map = Map('The variation of GDP',width = 1200,height = 600)
    map.add('GDP', attr, value, maptype = 'world', is_visualmap = True,
           visual_text_color = '#000', is_map_symbol_show=False,
           visual_range = [min_gdp, max_gdp],
           visual_range_text = ['%.3f'%min_gdp, '%.3f'%max_gdp],
           visual_top = 'center')
    timeline.add(map, col_name)
timeline.render()
```

# 数据对数变换后20年动态变换

```python
#取对数
def log(num):
    return np.log(num)
```

```python
for col_name in col_names[1:]:
    value = list(gdp[col_name])
    deal_value = []
    for val in value:
        deal_value.append(log(val))
    value = deal_value
    attr = list(gdp.country)
    map = Map('The variation of GDP',width = 1200,height = 600)
    map.add('GDP', attr, value, maptype = 'world', is_visualmap = True,
           visual_text_color = '#000', is_map_symbol_show=False,
           visual_range = [log(min_gdp), log(max_gdp)],
           visual_range_text = ['%.3f'%log(min_gdp), '%.3f'%log(max_gdp)],
           visual_top = 'center')
    timeline.add(map, col_name)
timeline.render()
```

### 取对数后，各国之间的GDP差距变得很小，不易区分，故本题采取原始数据展现世界各国得GDP总量和动态变化
